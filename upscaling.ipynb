```python
file_path = "/usr/local/lib/python3.11/dist-packages/basicsr/data/degradations.py"

with open(file_path, "r") as f:
    code = f.read()

# Replace the old (broken) import with the working one
patched_code = code.replace(
    "from torchvision.transforms.functional_tensor import rgb_to_grayscale",
    "from torchvision.transforms.functional import rgb_to_grayscale"
)

with open(file_path, "w") as f:
    f.write(patched_code)

print("✅ Patch applied to basicsr/data/degradations.py")
```

```python
import os
from PIL import Image

input_dir = "/kaggle/working/Real-ESRGAN/inputs"
clean_dir = "/kaggle/working/Real-ESRGAN/clean_inputs"

os.makedirs(clean_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    try:
        filepath = os.path.join(input_dir, filename)
        with Image.open(filepath) as img:
            if img.mode == "RGB":
                img.save(os.path.join(clean_dir, filename))
    except Exception as e:
        print(f"Skipped {filename}: {e}")
```
